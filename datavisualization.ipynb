```python
import pandas as pd
```

```python
df = pd.read_csv("recent-grads.csv")
```

```python
df.head(5)
```



---


Dataset contains some columns related to the earnings of graduates in each major:

"Median" is the median earnings of full-time, year-round workers.



"P25th" is the 25th percentile of earnings.



"P75th" is the 75th percentile of earnings.


"Rank" is the major’s rank by median earnings.








```python
df.plot(x="Rank", y=["P25th", "Median", "P75th"])
```



---


Based on the median infromation we get from the dataset we will compare majors and see which one has the highest salary.

```python
top_5 = df.sort_values(by="Median", ascending=False).head()
```

```python
top_5.plot(x="Major", y="Median", kind="bar", rot=10, fontsize=8)
```

This plot shows that the median salary of petroleum engineering majors is more than $20,000 higher than the rest.


---



```python
df.plot(x="Median", y="Unemployment_rate", kind="scatter")
```

No significant correlation between the earnings and unemployment rate.


---



```python
cat_totals = df.groupby("Major_category")["Total"].sum().sort_values()
```

```python
cat_totals
```

```python
cat_totals.plot(kind="barh", fontsize=10)
```

Business is the most popular major category. While humanities and liberal arts is the clear second, the rest of the fields are more similar in popularity.
